```python
from vg_mantra import build_feature_table
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
light=pd.read_csv("../res/transient_lightcurves_v2_filtered.csv")
light.set_index("MJD",inplace=True)
labels=pd.read_csv("../res/transient_labels2.csv")
```

```python
feats = build_feature_table(
    labels, light,
    min_instances=100,     # umbral mínimo de puntos por LC
    max_points=2000,       # recorte uniforme si hay series muy largas
    use_flux=True,        # o False → usa -Mag
    winsor=0.0,           # p.ej. 0.01 para recortar outliers
    whvg_scheme="delta_over_err",   # "delta" | "delta_over_err" | "zscore"
    use_delta_t_in_w=False          # True para dividir peso por (1+Δt)
)
feats.head()
```
